## **Challenge 4**

```python
#Vikram Nadathur May 26th 2026
## Print a customer receipt with 5 different products and very different prices as a table
# 1. Store products and prices in a list of tuples
products = [
    ('Pencil', 0.99),
    ('Notebook', 2.49),
    ('Backpack', 29.99),
    ('Calculator', 15.5),
    ('USB Drive', 8.75)
]

# 2. Print the header
print('Customer Receipt')
print('-' * 30)
print(f'{"Item":15}{"Price":>15}')
print('-' * 30)

# 3. Loop through the list and print each item dynamically
for item, price in products:
    print(f'{item:15}${price:>14.2f}')

print('-' * 30)

# 4. Use sum() to calculate the total automatically
total = sum(price for item, price in products)
print(f'{"Total":15}${total:>14.2f}')
```

```python
#Vikram Nadathur May 26th 2026
# Challenge: Write a program that reads a file and prints only the even numbers
with open('numbers.txt', 'r') as file:
    content = file.read() # Read the whole content as one string
    numbers_as_strings = content.split('\\n') # Split by the literal '\n' string

    for num_str in numbers_as_strings:
        cleaned_num_str = num_str.strip()
        if cleaned_num_str: # Ensure it's not an empty string
            number = int(cleaned_num_str)
            if number % 2 == 0:
                print(number)
```

# Slide 13: Sentinels

```python
# Using a sentinel value
total = 0
num = int(input("Enter a number (-1 to quit): "))
while num != -1:
    total += num
    num = int(input("Enter a number (-1 to quit): "))
print(f"Total: {total}")
```
